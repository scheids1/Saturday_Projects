```python
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline
```

### Data Source

https://www.lendingclub.com/info/download-data.action (Loan data 2016 Q1) 

```python
df = pd.read_csv("LoanStats_2016Q1.csv", skiprows=1, skipfooter=2, engine='python')
print(df.shape)
df.head()
```

## Data Cleaning

```python
# Dropping all columns that are all NaN
df = df.dropna(axis=1, how='all')
print(df.shape)
df.head()
```

```python
list(df.columns)
```

```python
df.dtypes
```

```python
# Convert interest rate to a float -- not a string 
df['int_rate'] = df['int_rate'].str.strip('%').astype(float)

# Convert term to an int -- not a string 
df['term'] = df['term'].str.strip('months').astype(int)
```

```python
# df.corr().abs().unstack().sort_values(ascending=False)
```

## Plotting & Styles

As this dataset has a wealth of variables, I decided to do some fun plotting and play around with features. 

A useful guide to matplotlib: https://realpython.com/python-matplotlib-guide/

```python
# Looking at the list of available styles
plt.style.available
```

**Let's look at the same data plotted using a few different styles. Fivethirtyeight, seaborn, and ggplot all tend to be popular options. **

Here, the style is set temporarily, not globally, for each cell/plot. To set it globally, call **plt.style.use("fivethirtyeight")**

```python
fig = plt.figure(figsize=(20,16))
ax1 = plt.subplot(221)
ax1.hist(df['loan_amnt'], bins=40)
plt.title("Loan Amounts")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
ax2 = plt.subplot(222)
ax2.hist(df['int_rate'], bins=30)
plt.title("Interest Rates")
plt.xlabel("Interest Rate")
plt.ylabel("Frequency")
ax3 = plt.subplot(223)
ax3.scatter(df['loan_amnt'], df['installment'], s=0.25)
plt.title("Loan Amount vs Installment Size")
plt.xlabel("Loan Amount")
plt.ylabel("Installment Amount")
ax4 = plt.subplot(224)
ax4.scatter(df['loan_amnt'], df['annual_inc'])
plt.title("Loan Amounts versus Annual Income")
plt.xlabel("Loan Amount")
plt.ylabel("Annual Income")
plt.show()
```
